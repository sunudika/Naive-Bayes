```python
import pandas as pd
from math import log
```

<b>pandas</b> buat manipulasi <i>data frame</i><br>
<b>math</b> buat ambil/melakukan operasi logaritma

```python
data = pd.read_csv("heart.csv")
data
```

Nilai pada variabel oldpeak kami ubah menjadi bilangan bulat dengan mengalikannya dengan 10 agar memudahkan proses-proses selanjutnya.

```python
data["oldpeak"]=data["oldpeak"]*10
data = data.astype('int64')
data
```

Inti dari praproses pada <i>notebook</i> ini adalah mengubah data <i>continous</i>/non-kategorikal pada variabel <b>Age</b>, <b>trestbps</b>, <b>chol</b>, <b>thalach</b>, dan <b>oldpeak</b> menjadi data kategorikal. Pertama-tama kita harus menentukan berapa banyak kelas/kategori yang dapat dibentuk melalui <i>dataset</i> tersebut, dengan rumus : <b>1 + 3,3 x log(banyak data atau baris)</b><br>

```python
jum_kls = 1+3.3*log(len(data), 10)
jum_kls
```

Selanjutnya adalah menentukan rentang nilai kelas/kategori pada variabel Age, trestbps, chol, thalach, dan oldpeak.

```python
def renNil(jar):
    if (((max(jar)-min(jar))/jum_kls) - int((max(jar)-min(jar))/jum_kls) > 0.5):
        return int((max(jar)-min(jar))/jum_kls)+1
    else:
        return int((max(jar)-min(jar))/jum_kls)
```

Rentang nilai tiap kelas/kategori pada kolom Age, trestbps, chol, thalach, dan oldpeak yang didapat adalah sebagai berikut :

```python
print(renNil(data["Age"]),
      renNil(data["trestbps"]),
      renNil(data["chol"]),
      renNil(data["thalach"]),
      renNil(data["oldpeak"]))
```

Maksudnya seperti ini, mari kita jadikan kolom Age sebagai contoh. Nilai terkecil dari kolom Age adalah 29(tahun) dan nilai terbesar dari kolom tersebut adalah 77(tahun). Dari 29 sampai 77 tersebut terbentuk 9.188760674057605 kategori(sengaja tidak kami bulatkan) di mana tiap kategori memiliki 5 rentang nilai, dengan rincian sebagai berikut :<br><br>

Umur 44-48 tahun masuk ke dalam kelas <b>3</b><br>
Umur 49-53 tahun masuk ke dalam kelas <b>4</b><br>
Umur 54-58 tahun masuk ke dalam kelas <b>5</b><br>
Umur 59-63 tahun masuk ke dalam kelas <b>6</b><br>
Umur 64-68 tahun masuk ke dalam kelas <b>7</b><br>
Umur 69-73 tahun masuk ke dalam kelas <b>8</b><br>
Umur 74-77 tahun masuk ke dalam kelas <b>9</b><br><br>
Hal ini berlaku juga pada variabel/kolom trestbps, chol, thalach, dan oldpeak. Banyaknya kelas/kategori adalah 9.188760674057605 yang jika dibulatkan menjadi 9. Kami sengaja tidak membulatkannya karena jika dibulatkan maka ini akan berpengaruh terhadap banyaknya jumlah anggota pada kelas/kategori tiap kolom dan pada akhirnya hanya terdapat 9 kelas saja. Padahal sebenarnya ada lebih dari 9 kelas. Hal ini kami lakukan untuk menampung nilai yang tersisa pada 0.188760674057605 sebagai satu kelas tambahan. Sehingga total dari banyaknya kelas keseluruhan adalah 10.

```python
list_var = [data["Age"], data["trestbps"], data["chol"], data["thalach"], data["oldpeak"]]
ren_var = [5,12,48,14,7]
baru = [[],[],[],[],[]]
```

Proses mengubah nilai :

```python
for n in range(len(list_var)):
    for i in list_var[n]:
        if(i <= min(list_var[n])+ren_var[n]):
            baru[n].append(0)
        elif(i <= min(list_var[n])+(ren_var[n]*2-1) and i >= min(list_var[n])+ren_var[n]):
            baru[n].append(1)
        elif(i <= min(list_var[n])+(ren_var[n]*3-1) and i >= min(list_var[n])+ren_var[n]*2):
            baru[n].append(2)
        elif(i <= min(list_var[n])+(ren_var[n]*4-1) and i >= min(list_var[n])+ren_var[n]*3):
            baru[n].append(3)
        elif(i <= min(list_var[n])+(ren_var[n]*5-1) and i >= min(list_var[n])+ren_var[n]*4):
            baru[n].append(4)
        elif(i <= min(list_var[n])+(ren_var[n]*6-1) and i >= min(list_var[n])+ren_var[n]*5):
            baru[n].append(5)
        elif(i <= min(list_var[n])+(ren_var[n]*7-1) and i >= min(list_var[n])+ren_var[n]*6):
            baru[n].append(6)
        elif(i <= min(list_var[n])+(ren_var[n]*8-1) and i >= min(list_var[n])+ren_var[n]*7):
            baru[n].append(7)
        elif(i <= min(list_var[n])+(ren_var[n]*9-1) and i >= min(list_var[n])+ren_var[n]*8):
            baru[n].append(8)
        elif(i <= min(list_var[n])+(ren_var[n]*10-1) and i >= min(list_var[n])+ren_var[n]*9):
            baru[n].append(9)
```

Masukkan nilai baru ke dalam variabel non kategorikal

```python
data["Age"]      = baru[0]
data["trestbps"] = baru[1]
data["chol"]     = baru[2]
data["thalach"]  = baru[3]
data["oldpeak"]  = baru[4]
```
