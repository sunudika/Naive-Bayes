### 2. Visualisasi data

<strong>2.1 Hubungan antara variabel Age, trestbps, chol, thalach, & oldpeak

```python
plt.style.use('seaborn')
sns.pairplot(data=data[['Age', 'trestbps', 'chol', 'thalach', 'oldpeak']])
```

Age = Umur pasien<br>
trestbps = Tekanan darah saat pasien tidak berolahraga(beraktivitas normal)<br>
chol = Kolestrol pasien<br>
thalach = Detak jantung paling banyak/menit<br>
oldpeak = Aktifitas listrik jantung(st depression => st segment)<br>

<strong>2.2 Sebaran gender koresponden

```python
sns.countplot(x=data.Gender)
```

1 = laki-laki<br>0 = perempuan<br>
Jumlah pasien laki-laki lebih banyak dua kali lipat dari perempuan.

<strong>2.3 Sebaran thalium pasien

```python
data['thal'].value_counts().plot(kind='pie')
```

3 : Normal<br>6 : Pasti terdeteksi<br>7 : Kadang terdeteksi

<strong>2.4 Sebaran jumlah pembuluh utama pada jantung pasien

```python
sns.countplot(y=data.ca)
```

<strong>2.5 Sebaran <i>fasting blood sugar</i>

fbs merupakan variabel yang merepresentasikan kadar gula darah pada pasien

```python
sns.countplot(x=data.fbs)
```

0 = Kurang dari 120<br>1 = Lebih dari 120

<strong>2.6 Sebaran data pada variabel slope

slope merupakan variabel yang merepresentasikan grafik aktifitas listrik pada jantung

```python
data['slope'].value_counts().plot(kind='pie')
```

1 : cenderung miring ke atas<br>2 : datar<br>3 : cenderung miring ke bawah

<strong>2.7 Sebaran data pada variabel restecg

restecg adalah hasil elektrokardiografi

```python
sns.countplot(y=data.restecg)
```

0 = normal<br>1 = memiliki gelombang ST-T yang tidak normal<br>2 = menunjukkan kemungkinan hipertrofi ventrikel kiri


<strong>2.8 Sebaran pada variabel cp

cp adalah variabel yang merepresentasikan tipe nyeri dada/<i>chest pain</i>

```python
sns.countplot(x=data.cp)
```

1 : typical angina(nyeri dada yang berhubungan dengan jantung)<br>2 : atypical angina(nyeri dada yang tidak berhubungan langsung dengan jantung)<br>3 : non-anginal pain(nyeri dada yang tidak berhubungan dengan jantung)<br>4 : asymptomatic(tidak menyebabkan adanya penyakit)

<strong>2.9 Sebaran pada variabel exang

exang merupakan variabel yang merepresentasikan apakah pasien terkena angina saat berolahraga

```python
data['exang'].value_counts().plot(kind='pie')
```

1 = Iya<br>0 = Tidak

<strong>2.10 Sebaran pada variabel num

num merupakan variabel yang merepresentasikan tingkat resiko terkena penyakit jantung

```python
sns.countplot(y=data.num)
```

0 : Tidak ada<br>1 : Rendah<br>2 : Lumayan beresiko<br>3 : Beresiko Tinggi<br>4 : Pasti memiliki penyakit jantung
